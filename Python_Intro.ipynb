```python
# DATA TYPES IN PYTHON
# There are more that aren't covered here like complex, frozenset, bytes etc.
n = 5 # int
print(n) # type() calls the data type of object
n = 1.3 # float
print(n)
n = "Hello World!" # str
print(n)
n = True # bool
print(n)
n = ["water", "fire", "air"] # list (0 indexed)
print(n)
n = ("water", "fire", "air") # tuple (0 indexed)
print(n)
n = {"bob": "smith", "michael": "david", "george": "washington"} # dict
print(n["bob"])
print(n)
n = {"apple","apple","banana","coconut","banana"} # set
print(n)
n = None # None datatype
print(n)
```

```python
# Lists are the most important and widely used data structure
a = [1,9,7,3]

a.append(2) # appends to the end of list
print(a)

a.insert(2,"happy") # insert(index, value)
print(a)

a.remove(3) # remove(index)
print(a)

popped = a.pop(1) # pop(index)
print(a)
print(popped)

a = [1,2,3,4,5]
print(a[2::]) # List splicing arr[start:stop:step]
print(a[::-1])
print(a[::2]) #odd indexes

# slicing is shallow copy, use .copy() for deep copy, especially with mutable objects
```

In the following coding segment, try to create a function that prints what day of the week it is based on an integer parameter corresponding to the day (i.e. Sunday = 1, Monday = 2, etc.). **The function should return a string** saying "The day is _____" with the blank being the corresponding day.

Check for invalid parameters (outside the range of 1-7, inclusive and null), and print an error message that says "Invalid day".

```python
def days_of_the_week(daysNum):
  match daysNum:
    case 1:
      print("The day is Sunday")
    case 2:
      print("The day is Monday")
    case 3:
      print("The day is Tuesday")
    case 4:
      print("The day is Wednesday")
    case 5:
      print("The day is Thursday")
    case 6:
      print("The day is Friday")
    case 7:
      print("The day is Saturday")
    case _:
      print("Invalid day")

days_of_the_week(1)
```

```python
days_of_the_week(-1)
```

```python
def get_day_of_week(day_num):
  days_map = {
      1: "Sunday",
      2: "Monday",
      3: "Tuesday",
      4: "Wednesday",
      5: "Thursday",
      6: "Friday",
      7: "Saturday"
  }

  if day_num in days_map:
    return f"The day is {days_map[day_num]}"
  elif day_num is None:
    return "Invalid day: Input cannot be None"
  else:
    return "Invalid day: Please enter a number between 1 and 7"

# Test cases
print(get_day_of_week(1))
print(get_day_of_week(4))
print(get_day_of_week(7))
print(get_day_of_week(0))
print(get_day_of_week(8))
print(get_day_of_week(None))
```
